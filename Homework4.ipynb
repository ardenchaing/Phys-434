```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
# 0:
t = np.linspace(0,np.pi*100,np.int(1e5))
N = 10+np.sin(t)
bkgd = stats.norm.rvs(size=np.int(1e5))*np.sqrt(N)+N
```

```python
# 1.a: 
scatterPlot = t[0:1000]
x = bkgd[0:1000]
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title("First 1k Pointsin Sinusoidal Data", fontsize=20)
ax.set_xlabel("t", fontsize=18)
ax.set_ylabel("Background", fontsize=18)
plt.scatter(scatterPlot, x)
plt.tick_params(labelsize = 16)
plt.show()
```

Problem 1.a: To plot a small section of the data set, I placed the wanted values into an array. Then I made a scatterplot of the array, which consists of the first 1k of the 100k data points given to me.

```python
# 1.b: 
def gen_scatter_plot(t, background, start, stop, step) :
    new_t = t[start:stop:step]
    new_background = background[start:stop:step]
    scatterPlot = t[0:1000]
    fig, ax = plt.subplots(1, 1, figsize=(10,10))
    ax.set_title("Graph from Point " + str(start) + " to " + str(stop) + " with Steps of " + str(step), fontsize=20)
    ax.set_xlabel("t", fontsize=18)
    ax.set_ylabel("Background", fontsize=18)
    plt.scatter(new_t, new_background)
    plt.tick_params(labelsize = 16)
    plt.show()
```

Problem 1.b: Above we can chop off and graph specific sections from the data set depending on the beginning and ending value. Also we do have the ability to ommit certain data points by only plotting certain values with our step variable. 

```python
gen_scatter_plot(t, bkgd, 1000, 3000, 1)
```

Above I have created a plot of the data points from 1k to 3k using the generalize code above. Plotting every single point (continuous). And below I have created another plot for points in the middle of the data set, also a continuous.

```python
gen_scatter_plot(t, bkgd, 50000, 52000, 1)
```

```python
# 1.c:
gen_scatter_plot(t, bkgd, 0, 100000, 100)
```

Problem 1.c: We want to plot the whole data set from 0 to 100k, but only every 100th point. We did this by setting our step variable to 100. 

```python
# 2.a:
binSize = 100
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title("2D Histogram Showcasing Central Density of a Data Set", fontsize=20)
ax.set_xlabel("t", fontsize=18)
ax.set_ylabel("Background", fontsize=18)
plt.tick_params(labelsize = 16)
plt.hist2d(t,bkgd,density=True,bins=binSize)
plt.colorbar()
plt.show()
```

Problem 2.a: Above is a 2d histogram plot showcasing the central density of the data points for values t and the corresponding background, which is sinusoidal. 

Problem 2b: In this plot, we can use the color bar to identify which areas are more dense than others. Looking at the yellow section correlates that there are a lot of points from either dataset in that location. And as you go down the bar, it indicates the decreasing density of data points, where purple is the area where there are little to no points from either dataset in that location. Now this is beneficial because the 2d histogram density plot is able to break up specific areas to give meaningful information. Below I graphed the entire data set in a scatterplot and it just looks like a blob. Everything blends together and it's difficult to decern any relationship. 

```python
gen_scatter_plot(t, bkgd, 0, 100000, 1)
```

```python
# 3.a: 
t_folded = t % (2*np.pi)
gen_scatter_plot(t_folded, bkgd, 0, 100000, 1)
```

Problem 3.a: Reusing the scatter plot function in problem 1, I created a new time variable which folds at 2pi and plotted the new full data set above. We can see that it has a sinusoidal shape which is due to the dataset being periodic. 

```python
# 3.b:
fig, ax = plt.subplots(1, 1, figsize=(10,10))
ax.set_title("2D Histogram Showcasing Central Density of a Data Set", fontsize=20)
ax.set_xlabel("t", fontsize=18)
ax.set_ylabel("Background", fontsize=18)
plt.tick_params(labelsize = 16)
plt.hist2d(t_folded,bkgd,density=True,bins=binSize)
plt.colorbar()
plt.show()
```
