Problem 1A-B:

```python
print (stats.norm.cdf(0))
print (stats.norm.cdf(0.25))
print (stats.norm.cdf(0.5))
print (stats.norm.cdf(0.75))
print (stats.norm.cdf(1))
print (stats.norm.cdf(2))
print (stats.norm.cdf(3))
```

Above we are double checking the values of sigma and comparing it against the z-table. We can see that as sigma increases so does the probablilty, and within two standard deviations of the mean it covers around 98% of the data. This follows exactly what one'd learn about a normal Gaussian distrubution in probablity class. 

Problem 1C:

```python
print (stats.norm.ppf(0.5))
print (stats.norm.ppf(0.84134))
print (stats.norm.ppf(0.97725))
print (stats.norm.ppf(0.99865))
```

Above I wanted to verify the sigma values associated with percentages. As shown above the relationship of the probablity to sigma value isn't linear but rather expotential. An example showcasing that is the difference between 0-1 sigma being 34% and 2-3 being 2%.

Problem 2A-B:

```python
d = stats.expon.rvs(loc = 0, scale = 1, size = 100000)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True, label='Histogram')
plt.tick_params(labelsize = 20)
plt.xlim([0,5])
x = np.linspace(0,5,1000)
ax.plot(x,stats.expon.pdf(x,loc = 0, scale = 1), linewidth = 8,alpha = 0.7, label = 'Exponential')
plt.xlabel('Value', fontsize = 27) 
plt.ylabel('Probablity Density', fontsize = 27)
plt.title('Exponential Distrubution w/ 100k Samples', fontsize = 30)
plt.legend(fontsize=25)
plt.show()
```

I decided to choose an exponential function. As you can see the probablity density decreaess drastically as the values of x increase.
