Problem 2.1.A: A binomial distribution is a "discrete probablity distribution of the number of successes in a sequence of n independent experiments, each asking a yes-no question, and each with its own Boolean-valued outcome." The probability mass function (PMF) is [n(factorial)] / [k(factorial)*(n-k)(factorial)] * p^k * (1-p)^(n-k). Where p is the probablity of success and n is the amount of trials. https://en.wikipedia.org/wiki/Binomial_distribution

```python
# 2.1.B:
n = 10
p = 0.5
# defining list of r values 
r_values = list(range(n + 1)) 
# list of pmf values 
dist = [binom.pmf(r, n, p) for r in r_values] 
# plotting the graph
plt.step(r_values, dist)
plt.xlabel('X', fontsize = 27) 
plt.ylabel('P(x)', fontsize = 27)
plt.title('Binomial Distribution w/ n = 10, p = .5', fontsize = 30)
plt.show()
```

Above shows with n trials and the probablity of success being 50% what is the spread across those trials. You can see since it's a 50/50 shot, the spread is mirrored across the medium which is 5. 

```python
n = 10
p = .3
# defining list of r values 
r_values = list(range(n + 1)) 
# list of pmf values 
dist = [binom.pmf(r, n, p) for r in r_values ] 
# plotting the graph  
plt.step(r_values, dist) 
plt.xlabel('X', fontsize = 27) 
plt.ylabel('P(x)', fontsize = 27)
plt.title('Binomial Distribution w/ n = 10, p = .3', fontsize = 30)
plt.show()
```

Above you see that with 10 trials and the probablity of success being 30%, what's the spread across the field. It seems like the mode of each trial is n*p. 

```python
n = 10
p = .9
# defining list of r values 
r_values = list(range(n + 1)) 
# list of pmf values 
dist = [binom.pmf(r, n, p) for r in r_values ] 
# plotting the graph  
plt.step(r_values, dist) 
plt.xlabel('X', fontsize = 27) 
plt.ylabel('P(x)', fontsize = 27)
plt.title('Binomial Distribution w/ n = 10, p = .9', fontsize = 30)
plt.show()
```

Seen in the code above with a binomial distrbution, it's almost impossible to succeed less than five times in a 10 trial grouping.
