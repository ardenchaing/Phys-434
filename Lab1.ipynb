Problem 1A-B:

```python
print (stats.norm.cdf(0))
print (stats.norm.cdf(1))
print (stats.norm.cdf(2))
print (stats.norm.cdf(3))
```

Above we are calculating the probablity that a value is between the left end bound (negative infinity) and the associated z value in a standard normal distribution. Looking at a typical bell curve graph we can see that the data spread it split equally between the mean. That is further confirmed with the first value which asks what is the probablity that a value is between negative infinity and the mean plus zero standard deviation (which is just the mean). It's 50%. Also we can further calidate this claim since we know that within one standard deviation of the mean, it encapsulates 68% of the data. And within two standard deviations of the mean, it encapsulates 95% of the data. Using this knowledge we can say that the probablity that a value is between zero and two standard deviations above the mean is (1/2)(.95) + (.50) = 97.5% which is pretty close to the value we calculated above.

Problem 2A-B:

```python
d = stats.expon.rvs(loc = 0, scale = 1, size = 100000)
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True, label='Histogram')
plt.tick_params(labelsize = 20)
plt.xlim([0,5])
x = np.linspace(0,5,1000)
ax.plot(x,stats.expon.pdf(x,loc = 0, scale = 1), linewidth = 8,alpha = 0.7, label = 'Exponential')
plt.xlabel('Value', fontsize = 27) 
plt.ylabel('Probablity Density', fontsize = 27)
plt.title('Exponential Distrubution w/ 100k Samples', fontsize = 30)
plt.legend(fontsize=25)
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.hist(d,50, density=True, label='Histogram')
plt.yscale('log')
plt.tick_params(labelsize = 20)
plt.xlim([0,5])
x = np.linspace(0,5,1000)
ax.plot(x,stats.norm.pdf(x,loc = 0, scale = 1),linewidth = 8,alpha = 0.7, label='Exponential')
plt.xlabel('Value', fontsize = 27) 
plt.ylabel('Log()', fontsize = 27)
plt.title('Exponential Distrubution w/ 100k Samples', fontsize = 30)
plt.legend(fontsize=25)
plt.show()
```

I decided to choose an exponential function. As you can see the probablity density decreaess drastically as the values of x increase.
