```python
# 2.A: 
x = np.linspace(0,100,1001)
xc = np.linspace(0,200,2001)
ex1 = stats.expon.pdf(x)
ex2 = stats.expon.pdf(x)
both = np.convolve(ex1,ex2)
fig, ax = plt.subplots(1)
ax.plot(x,ex1, label='one expon distribution')
ax.plot(xc, both, label='sum of expon distribution')
ax.set_xlim([-1,20])
plt.legend(loc='best')
plt.show()
```
